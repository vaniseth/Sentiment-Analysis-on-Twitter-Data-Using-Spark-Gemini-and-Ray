```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Extract all the hashtags and URLs in the tweets 

```python
# Read JSON file from data folder
data = pd.read_json('data/out.json', lines=True)

# Display the table
data.head()
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = data.isnull().mean() * 100

# Plot the missing value percentage
missing_percentage.plot(kind='bar', figsize=(12, 6), title='Missing Value Percentage by Column')
plt.ylabel('Percentage')
plt.show()
```

```python
# Get columns with more than 50% missing values
columns_to_drop = missing_percentage[missing_percentage > 50].index

# Drop those columns from the dataframe
data_cleaned = data.drop(columns=columns_to_drop)

# Display the cleaned dataframe
data_cleaned.head()
```

```python
# Extract hashtags and URLs from the tweets
data_cleaned['hashtags'] = data_cleaned['entities'].apply(lambda x: [hashtag['text'] for hashtag in x['hashtags']])
data_cleaned['urls'] = data_cleaned['entities'].apply(lambda x: [url['expanded_url'] for url in x['urls']])

# Display the hashtags and URLs
data_cleaned[['hashtags', 'urls']].head()
```

```python
# Drop rows where 'hashtags' or 'urls' columns contain empty lists
hash_url_df = data_cleaned[['hashtags', 'urls']]
print(hash_url_df.shape)
hash_url_df = hash_url_df[(hash_url_df['hashtags'].str.len() > 0) & (hash_url_df['urls'].str.len() > 0)]

# Display the cleaned dataframe
print(hash_url_df.shape)
hash_url_df.head()
```

```python
# Open a text file in write mode
with open('hashtags_urls.txt', 'w') as file:
    # Write hashtags to the file
    file.write('Hashtags:\n')
    for hashtags in hash_url_df['hashtags']:
        for hashtag in hashtags:
            file.write(f"{hashtag}\n")
    
    # Write URLs to the file
    file.write('\nURLs:\n')
    for urls in hash_url_df['urls']:
        for url in urls:
            file.write(f"{url}\n")
```

```python
from collections import Counter

# Initialize a Counter object to store word counts
word_counts = Counter()

# Open the file and read it line by line
with open('hashtags_urls.txt', 'r') as file:
    for line in file:
        # Split each line into words and update the counter
        words = line.strip().split()
        word_counts.update(words)

# Display the word counts
for word, count in word_counts.items():
    print(f"{word}: {count}")
```
